Oh uh!  The unthinkable has occurred and a zombie outbreak has occurred on Square Island!  This island is 100 square miles and evenly populated by 100 citizens.  Can Square Island's once peaceful populace hold out for 10 days until help arrives?  Let's find out!

To model the Square Island's predicament, we are going to create an agent based model.  This model will include the island's 100 inhabitants as actors that will interact with each other as the outbreak spreads.  The individual inhabitants of the island, refered to as 'citizens' from now on, will have three possible conditions: Human, Zombie, and Dead.  All the citizens begin as humans and then a group immediately will become zombies.  These zombies will be able to convert adjacent humans into zombies, while humans will be able to make neighboring zombies dead.  We will iterate this process ten times and see how many humans make it to the end!

```python
#First, let's load some packages, including mesa

import random
import mesa
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

from mesa import Model, Agent
from mesa.time import RandomActivation
from mesa.space import Grid
from mesa.datacollection import DataCollector
from mesa.batchrunner import BatchRunner
```

```python
#Below we define the agent aspects of the model
#The citizens of Square Island can have three states: Human, Zombie, or Dead

class CitizenCell(Agent):

    def __init__(self, model, pos):
     
        super().__init__(pos, model)
        self.pos = pos
        self.unique_id = pos
        self.condition = "Human"
        
        
#Two things can happen each day (timestep), zombies can infect neighboring humans and humans can defend themselves and kill neighboring zombies   
        
    def step(self):
        
        #Zombies infect their neighbors with a 50% probability
        if self.condition == "Zombie":
            neighbors = self.model.grid.get_neighbors(self.pos, moore=False)
            for neighbor in neighbors:
                if neighbor.condition == "Human" and random.random()>=.5:
                    neighbor.condition = "Zombie"
            self.condition = "Zombie"
            
        #Humans only defend themselves with a 10% probability    
        if self.condition == "Human":
            neighbors = self.model.grid.get_neighbors(self.pos, moore=False)
            for neighbor in neighbors:
                if neighbor.condition == "Zombie" and random.random()>=.9:
                    neighbor.condition = "Dead"
            self.condition = "Human"
```

```python
# Next, we define our model
class Outbreak(Model):
   
    def __init__(self, height, width, density): #What does the self arguement do? Do I define it in the next cell?
        '''
        Create a new model.
        
        Args:
            height, width: The size of the grid to model
            density: What fraction of grid cells have a tree in them.
        '''
        # Initialize model parameters
        self.height = height
        self.width = width
        self.density = density
        
        # Set up model objects
        self.schedule = RandomActivation(self)
        self.grid = Grid(height, width, torus=False)
        self.dc = DataCollector({"Human": lambda m: self.count_type(m, "Human"),
                                "Zombie": lambda m: self.count_type(m, "Zombie"),
                                "Dead": lambda m: self.count_type(m, "Dead")})
        
        # Place a citizen in each cell with Prob = density.  We will set this probability = 1 so that every grid location is filled.
        for x in range(self.width):
            for y in range(self.height):
                if random.random() < self.density:
                    # Create a citizen
                    new_citizen = CitizenCell(self, (x, y))
                    # Infect all humans in the first column.
                    if x == 0:
                        new_citizen.condition = "Zombie"
                    self.grid[y][x] = new_citizen
                    self.schedule.add(new_citizen)
        self.running = True
        
    def step(self):
        '''
        Advance the model by one step.
        '''
        self.schedule.step()
        self.dc.collect(self)
        # Halt if no more humans
        if self.count_type(self, "Human") == 0:
            self.running = False
            
    @staticmethod
    def count_type(model, citizen_condition):
        '''
        Helper method to count citizens in a given condition in a given model.
        '''
        count = 0
        for citizen in model.schedule.agents:
            if citizen.condition == citizen_condition:
                count += 1
        return count
```

```python
#Now we set the island size and set the island's density parameter, which in this case is 1, in order to populate it.
#The outbreak has begun!  Let's run this model
outbreak = Outbreak(10, 10, 1)
#This will limit the model to ten iterations
for i in range(10):
    outbreak.step()
#This will produce a graph of the number of humans, zombies, and dead throughout the ten steps    
results = outbreak.dc.get_model_vars_dataframe()
results.plot()
```

```python
#As we can see above, the humans are not doing so hot.  Even when the model is run multiple times, the humans rarely survive until the end
#However, the humans will probably have better luck on a bigger, but more sparesely populated, island.

#This island is 15x15, but only 3/4 of the positions are filled, limiting the number of potential interactions
outbreak = Outbreak(15, 15, .75)

for i in range(10):
    outbreak.step()

results = outbreak.dc.get_model_vars_dataframe()
results.plot()
```

Well, would you look at that!  The humans are surviving a lot better on the larger island, though the zombies are still getting a good percentage of them.  With a larger island, the zombies would need more iterations to convert all of the available humans.  Additionally, the lower density means that some citizens have fewer neighbors and thus a lower chance of being turned into a zombie.  I suppose that the take away here is that infections spread best in densly packed areas where there are more potential interactions between actors.
